# What is the average, median and mode for word length

```python
import numpy as np
file=open('ques2.txt','r+')
count1=0
wordslength=[]
for word in file.read().split(" "):
    word=word.strip(".,?:()\"")
    word=word.replace(".","")
    word=word.replace("'","")
    wordslength+=[len(word)]
wordslength.sort()
count2=np.bincount(wordslength)
print('MEAN    = ',np.mean(wordslength))
print('MEDIAN  = ',np.percentile(wordslength,50))
print('MODE    = ',np.argmax(count2))
```
